```python
import cv2
import matplotlib.pyplot as plt
from skimage.feature import hog
from skimage import data, exposure

# Read the image
img = cv2.imread("C:/Users/HP/Data_Mining/Standford Dogs Dataset/images/Images/n02112706-Brabancon_griffon/n02112706_2184.jpg")

# # Visualize the image
# plt.axis('off')
# plt.imshow(img, cmap=plt.cm.gray)
# plt.title('Input image')
```

```python
# Compute the HOG descriptors
fd, hog_image = hog(img, orientations=8, pixels_per_cell=(16, 16), cells_per_block=(1, 1), visualize=True, channel_axis=-1)

# Rescale the HOG descriptors for better display
hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))

# Visualize the HOG descriptors
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4), sharex=True, sharey=True)
ax1.axis('off')
ax1.imshow(img, cmap=plt.cm.gray)
ax1.set_title('Input image')
ax2.axis('off')
ax2.imshow(hog_image, cmap=plt.cm.gray)
ax2.set_title('Histogram of Oriented Gradients')
plt.show()

# Visualize the HOG descriptors
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4), sharex=True, sharey=True)
ax1.axis('off')
ax1.imshow(img, cmap=plt.cm.gray)
ax1.set_title('Input image')
ax2.axis('off')
ax2.imshow(hog_image_rescaled, cmap=plt.cm.gray)
ax2.set_title('Histogram of Oriented Gradients - Rescaled')
plt.show()
```
