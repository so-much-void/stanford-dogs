```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage import filters
```

```python
# Get Path of Images.
base_path = "C:/Users/HP/Data_Mining/Cropped/"
selected_folders = ['n02085936-Maltese_dog', 'n02112706-Brabancon_griffon', 'n02086910-papillon', 'n02088466-bloodhound']
image_names = {
    'n02085936-Maltese_dog': ['n02085936_2889-0.jpg', 'n02085936_3313-0.jpg'],
    'n02112706-Brabancon_griffon': ['n02112706_1180-0.jpg', 'n02112706_363-0.jpg'],
    'n02086910-papillon': ['n02086910_862-0.jpg', 'n02086910_7949-0.jpg'],
    'n02088466-bloodhound': ['n02088466_2908-0.jpg', 'n02088466_3032-0.jpg']
}

# List images from selected folders
selected_images = []
for folder in selected_folders:
    folder_path = os.path.join(base_path, folder)
    images = [os.path.join(folder_path, img) for img in image_names[folder]]
    selected_images.extend(images)

# Now perform operations on selected images
# rows_plot = 2  # Number of rows for subplot
# column_plot = len(selected_images) // rows_plot + 1
rows_plot = 2  # Number of rows for subplot
column_plot = 4

greyScaleImgIndex = 1
histGramindex = 1
edgeindex = 1

# Iterate over selected images
for imgPlot in selected_images:
    image = cv2.imread(imgPlot)
    greyImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Perform edge detection using Sobel filter
    edge_sobel = filters.sobel(greyImage)
    
    # Plot greyscale images
    plt.figure("GreyScale Images")
    plt.subplot(rows_plot, column_plot, greyScaleImgIndex)
    plt.title(greyScaleImgIndex)
    plt.imshow(greyImage, cmap='gray')
    greyScaleImgIndex += 1

    # Plot histograms
    plt.figure("GreyScale Images Histogram")
    plt.subplot(rows_plot, column_plot, histGramindex)
    plt.title(histGramindex)
    plt.xlabel("Pixel Intensity")
    plt.hist(greyImage.ravel(), 256, [0, 256])
    histGramindex += 1

    # Plot edge detection result using Sobel filter
    plt.figure("Sobel Edge Detection")
    plt.subplot(rows_plot, column_plot, edgeindex)
    plt.title(histGramindex)
    plt.imshow(edge_sobel, cmap=plt.cm.gray)
    edgeindex += 1

plt.tight_layout()
plt.show()
```
